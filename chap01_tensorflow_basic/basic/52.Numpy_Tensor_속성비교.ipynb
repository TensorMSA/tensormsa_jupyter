```python
import numpy as np
a = np.array([1,2,3,4,5])
print(np.ndim(a))
print(np.shape(a))
print(np.size(a))
```

```python
import tensorflow as tf
ta = tf.constant([1,2,3,4,5])

with tf.Session() as sess:
    print(sess.run(tf.rank(ta)))
    print(sess.run(tf.shape(ta)))
    print(sess.run(tf.size(ta)))
```
